#Importing libraries

```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
from tensorflow.keras.datasets import mnist



```

#Loading and Preprocessing data

```python
# Load and preprocess data
(x_train, y_train), (x_test, y_test) = mnist.load_data()

```

#Normalization

```python
#Normalizing the dataset

x_train_norm = x_train / 255.0
x_test_norm = x_test / 255.0
x_test_norm,x_test_norm
```

```python
print(x_train.shape)
print(x_test.shape)
#These are not in one hot code
```

```python
#The data has to be in one-Hot encode the training or testing  labels  data has to be converted into OHC
x_train_one_hot = tf.one_hot(y_train,depth = 10)
x_test_one_hot = tf.one_hot(y_test,depth = 10)

x_test_one_hot
```

```python
#Building the model

model_1 = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape = (28,28)),
    tf.keras.layers.Dense(100,activation ="relu"),
    tf.keras.layers.Dense(100,activation ="relu"),
    tf.keras.layers.Dense(10,activation= "softmax")
])

#Compile the model

model_1.compile(loss= tf.keras.losses.SparseCategoricalCrossentropy(),
              optimizer = tf.keras.optimizers.Adam(),
              metrics= ["accuracy"])

#Fit the model

non_norm_history = model_1.fit(x_train,y_train,epochs = 10,validation_data =(x_test,y_test))
```

#Building of model

```python
#Building the model normalized

model_2 = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape = (28,28)),
    tf.keras.layers.Dense(100,activation ="relu"),
    tf.keras.layers.Dense(100,activation ="relu"),
    tf.keras.layers.Dense(10,activation= "softmax")
])

#Compile the model

model_2.compile(loss= tf.keras.losses.SparseCategoricalCrossentropy(),
              optimizer = tf.keras.optimizers.Adam(),
              metrics= ["accuracy"])

#Fit the model

norm_history = model_2.fit(x_train_norm,y_train,epochs = 10,validation_data =(x_test_norm,y_test))
```

```python
#Plot the curves of normalized and non normalized for the accuracy
import pandas as pd

#Plot normalized
pd.DataFrame(norm_history.history).plot(title = "normalized data ")
#Plot non_normalized
pd.DataFrame(non_norm_history.history).plot(title = "Non-normalized data ")
```

#Evaluation

```python
#The accuracy of both the normalized data and non-normalized dta model

#For model_1 non_normalizd
loss1, accuracy1 = model_1.evaluate(x_test, y_test, verbose=0)


#The accuracy of both the normalized data and non-normalized dta model

#For model_1 non_normalizd
loss1, accuracy1 = model_1.evaluate(x_test, y_test, verbose=0)

#For model_2 normalized
loss, accuracy = model_2.evaluate(x_test, y_test, verbose=0)

print(f"the accuracy of model_1 is {accuracy1}")
print(f"the accuracy of model_2 is {accuracy}")
```
